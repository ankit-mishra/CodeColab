```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob

import librosa as lr
```

```python
# Set directory for source files
data_dir = './heartbeat_sounds/set_a'
audio_files = glob(data_dir + '/*.wav')
```

```python
print(len(audio_files))
audio, sfreq = lr.load(audio_files[7])
time = np.arange(0, len(audio)) / sfreq
```

```python
fig, axe = plt.subplots()
axe.plot(time, audio)
axe.set(xlabel = 'Time (s)', ylabel='Sound Amplitude')
plt.show()
```

```python
# Perform the reading through all the audio files
for file in range(0, len(audio_files),1):

    # Read in the first audio and create the time array (timeline)
    audio, sfreq = lr.load(audio_files[file])
    time = np.arange(0, len(audio)) / sfreq

    # Plot audio over time
    fig, axe = plt.subplots()
    axe.plot(time, audio)
    axe.set(xlabel = 'Time (s)', ylabel='Sound Amplitude')
    plt.show()
```
