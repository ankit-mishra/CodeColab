```
import json
import csv
from pyedflib import highlevel
from pyedflib import EdfReader

CSV_COLUMNS = ['label', 'dimension', 'sample_rate', 'physical_max', 'physical_min', 'digital_max', 'digital_min',
               'prefilter', 'transducer']
csv_filename = "patient_details.csv"
# read an edf file
signals, signal_headers, header = highlevel.read_edf('/content/drive/My Drive/Tensorflow/SC4001E0-PSG.edf')

print(
    '********************************************************************************************************************************',
    file=open("output.txt", "a"))
f = EdfReader('/content/drive/My Drive/Tensorflow/SC4001E0-PSG.edf')
# print("\nlibrary version: %s" % pyedflib.version.version)

print('********************** This is EDF Header ******************************************')

print("file duration: %i seconds" % f.file_duration)
print("startdate: %i-%i-%i" % (f.getStartdatetime().day, f.getStartdatetime().month, f.getStartdatetime().year))
print("starttime: %i:%02i:%02i" % (f.getStartdatetime().hour, f.getStartdatetime().minute, f.getStartdatetime().second))
print("patientcode: %s" % f.getPatientCode())
print("gender: %s" % f.getGender())
print("birthdate: %s" % f.getBirthdate())
print("patientname: %s" % f.getPatientName())
print("patient_additional: %s" % f.getPatientAdditional())
print("admincode: %s" % f.getAdmincode())
print("technician: %s" % f.getTechnician())
print("equipment: %s" % f.getEquipment())
print("recording_additional: %s" % f.getRecordingAdditional())
print("datarecord duration: %f seconds" % f.getFileDuration())
print("number of datarecords in the file: %i" % f.datarecords_in_file)
print("number of annotations in the file: %i" % f.annotations_in_file)

print('*********************** This is Signal Header *****************************************')
with open(csv_filename, 'w', newline="") as csvfile:
    count = 0
    for i in signal_headers:
        json_parsed = json.dumps(i)
        json_loaded = json.loads(json_parsed)
        print(json_loaded)
        try:
            writer = csv.DictWriter(csvfile, fieldnames=CSV_COLUMNS)
            if count == 0:
                writer.writeheader()
            writer.writerow(i)
            count += 1
        except IOError:
            print("I/O error")
    print(i)

print('*********************** This is the Signals *****************************************')

for i in signals:
    print(i)

print('****************************************************************')
for i in header:
    print(i, file=open("patient_csv.csv", "a"))

print('*********************** This is Signal Header *****************************************')
with open(csv_filename, 'w', newline="") as csvfile:
    count = 0
    for i in signal_headers:
        json_parsed = json.dumps(i)
        json_loaded = json.loads(json_parsed)
        print(json_loaded)
        try:
            writer = csv.DictWriter(csvfile, fieldnames=CSV_COLUMNS)
            if count == 0:
                writer.writeheader()
            writer.writerow(i)
            count += 1
        except IOError:
            print("I/O error")
    print(i)

print('*********************** This is the Signals *****************************************',
      file=open("output.txt", "a"))

for i in signals:
    print(i)

print('****************************************************************')
```
